```python
%matplotlib inline
import matplotlib.pyplot as plt

from scoupy.sedimentsizedistribution import LogScaleSedimentSizeDistribution

size_distribution = LogScaleSedimentSizeDistribution(5e-5, 0.25)
_ = plt.plot(*size_distribution.cdf())
```

```python
from scoupy.sedimentsample import SedimentSample

sample = SedimentSample(100, size_distribution=size_distribution)
fines_sample, sand_sample = sample.split()
```

```python
_ = plt.plot(*fines_sample.size_distribution().cdf())
```

```python
_ = plt.plot(*sand_sample.size_distribution().cdf())
```

```python
summed_sample = sand_sample.add(fines_sample)
_ = plt.plot(*summed_sample.size_distribution().cdf())
```

```python
summed_sample.concentration()
```

```python
import numpy as np

_, cdf = sample.size_distribution().cdf()
_, summed_cdf = summed_sample.size_distribution().cdf()

np.allclose(cdf, summed_cdf, atol=1e-5, rtol=0)
```
